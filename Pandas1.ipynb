```python
#Find batsman with most number of 4's and 6's in last five overs
temp_df=ipl[ipl['over']>15]
temp_df[(temp_df['batsman_runs']==4) | (temp_df['batsman_runs']==6) ]
temp_df.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1)
```

```python
#Create a function that can return the highest score of any batsman
def highest(batsman):
    temp_df=ipl[ipl['batsman']==batsman]
    return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
#Merging Joining and Concatenating
courses=pd.read_csv('courses.csv')
matches=pd.read_csv('matches.csv')
nov=pd.read_csv('reg-month1.csv')
dec=pd.read_csv('reg-month2.csv')
students=pd.read_csv('students.csv')
students.head(2)
courses
students
regs
```

```python
#pd.concat
regs=pd.concat([nov,dec],ignore_index=True)
#Multi_Index DataFrame
multi=pd.concat([nov,dec],keys=['Nov','Dec'])
multi.loc[('Nov',4)]
regs.head(2)
```

```python
# pd.concat([nov,dec],axis=1)
```

```python
pd.concat([nov,dec],axis=1).head() #Concatenating horizontally
```

```python
#Merge
#inner join
students.merge(regs,how='inner',on='student_id')
```

```python
#Left join
courses.merge(regs,how='left',on='course_id').tail(2)
```

```python
#June 16
#Outer join
students.merge(regs,how='outer',on='student_id').tail(1)
```

```python
#Find total revenue generated
courses.merge(regs,how='inner',on='course_id')['price'].sum()
```

```python
#Find month by month revenue
temp_df=pd.concat([nov,dec],keys=['Nov','Dec']).reset_index()
temp_df.merge(courses,on='course_id').groupby('level_0')['price'].sum()
```

```python
#Print the registration table cols-> name -> course-> price
regs.merge(students,on='student_id').merge(courses,on='course_id')[['name','course_name','price']].sample()
```

```python
#Plot bar chart for revenue/course
regs.merge(courses,how='inner',on='course_id').groupby('course_name')['price'].sum().plot(kind='bar')
```

```python
#find students who enrolled in both the months
import numpy as np
common_student_id=np.intersect1d(dec['student_id'],nov['student_id'])
students[students['student_id'].isin(common_student_id)].sample()
```

```python
#Find the course that got no enrollment
course_common=np.setdiff1d(courses['course_id'],regs['course_id'])
courses[courses['course_id'].isin(course_common)]
```

```python
#print students name -> partner name for all enrolled students
#Self Join
students.merge(students,how='inner',left_on='partner',right_on='student_id')[['name_x','name_y']].head(2)
```

```python
#Find top 3 students who did most number enrollment
regs.merge(students,on='student_id').groupby(['student_id','name'])['name'].count().sort_values(ascending=False).head(2)
```

```python
#Find the top students who spent most amount of money on courses
regs.merge(students,on='student_id').merge(courses,on='course_id').groupby(['student_id','name'])['price'].sum().sort_values(ascending=False).head(2)
```

```python
#Find top 3 stadiums with highest sixes/match ratio
temp=ipl.merge(matches,left_on='match_id',right_on='id')
six=temp[temp['batsman_runs']==6]
num_sixes=six.groupby('venue')['venue'].count()
num_matches=matches['venue'].value_counts()
(num_sixes/num_matches).sort_values(ascending=False).head(2)
```

```python
#Find orange cap holder of all season
temp.groupby(['season','batsman'])['batsman_runs'].sum().reset_index().sort_values('batsman_runs',ascending=False).drop_duplicates(subset=['season'],keep='first').sort_values('season').head(2)
```

```python
#Vectorized Strings Operations
df=pd.read_csv('titanic.csv')
df['Name'][df['Name'].str.len()==82].values[0]
#split ->get
df['LastName']=df['Name'].str.split(',').str.get(0)
df[['title','firstname']]=df['Name'].str.split(',').str.get(1).str.strip().str.split( ' ',n=1,expand=True)
df.sample()
```

```python
#replace
df['title']=df['title'].str.replace('Ms.','Miss.')
df['title']=df['title'].str.replace('Mlle.','Miss.')
```

```python
#Filtering
#Find lastnames with start and end char vowel
df[df['LastName'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')].sample()
```
